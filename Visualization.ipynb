```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
df = pd.read_csv(r"C:\Users\sksan\OneDrive\Desktop\CodeAlpha\quotes_dataset.csv")
df.head()
```

Top Authors

```python
plt.figure(figsize=(8,5))
df['Author'].value_counts().head(10).plot(
    kind='bar',
    color='skyblue',
    edgecolor='black'
)

plt.title("Top 10 Authors by Number of Quotes", fontsize=12)
plt.xlabel("Author")
plt.ylabel("Number of Quotes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Quote Length

```python
df['Quote_Length'] = df['Quote'].apply(len)

plt.figure(figsize=(8,5))
plt.hist(
    df['Quote_Length'],
    bins=20,
    color='orchid',
    edgecolor='black'
)

plt.title("Distribution of Quote Lengths", fontsize=12)
plt.xlabel("Quote Length")
plt.ylabel("Frequency")
plt.show()
```

Common Tags

```python
tags = df['Tags'].str.split(", ").explode()
top_tags = tags.value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(
    x=top_tags.values,
    y=top_tags.index,
    palette="viridis"
)

plt.title("Top 10 Most Common Tags", fontsize=12)
plt.xlabel("Frequency")
plt.ylabel("Tags")
plt.show()
```

Outliers

```python
plt.figure(figsize=(8,4))
sns.boxplot(
    x=df['Quote_Length'],
    color='lightgreen'
)

plt.title("Quote Length Distribution (Outliers)", fontsize=12)
plt.xlabel("Quote Length")
plt.show()
```

```python
plt.axvline(
    df['Quote_Length'].mean(),
    color='red',
    linestyle='--',
    label='Average Length'
)
plt.legend()
```

Category-Based Story 

```python
df['Quote_Type'] = pd.cut(
    df['Quote_Length'],
    bins=[0,80,150,1000],
    labels=["Short","Medium","Long"]
)

df['Quote_Type'].value_counts().plot(
    kind='bar',
    color=['lightblue','orange','green']
)
plt.show()
```
